```python
notas1 = '''81,60,72,24,15,91,12,70,29,42,16,3,35,67,10,57,11,69,12,77,13,86,48,65,51,41,87,43,10,87,91,15,44,85,73,37,42,95,18,7,74,60,9,65,93,63,74'''

notas2 = '''30,95,28,84,84,43,66,51,4,11,58,10,13,34,96,71,86,37,64,13,8,87,14,14,49,27,55,69,77,59,57,40,96,24,30,73,95,19,47,15,31,39,15,74,33,57,10'''

alumnos = """'Agustin','Alan','Andrés','Ariadna','Bautista','CAROLINA','CESAR','David','Diego','Dolores','DYLAN','ELIANA','Emanuel','Fabián','Facundo','Facundo','FEDERICO','FEDERICO','GONZALO','Gregorio','Ignacio','Jonathan','Jonathan','Jorge','JOSE','JUAN','Juan','Juan','Julian','Julieta','LAUTARO','Leonel','LUIS','Luis','Marcos','María','MATEO','Matias','Nicolás','NICOLÁS','Noelia','Pablo','Priscila','TOMAS','Tomás','Ulises', 'Yanina'"""



def prolijo(texto):
    texto=quitar_comas(texto)
    return texto.split()
```

La función prolijo toma el texto de nombres y le quita las comillas y comas innecesarias.

```python
def quitar_comas(texto):
    coma_comilla = "',"
    for letra in texto:
        if letra in coma_comilla:
            texto = texto.replace(letra, " ")
    return texto
```

La función quitar_comas, es la que recorre las palabras contenidas en la variable texto y con la función replace reemplaza las comas y comillas por el espacio para limpiar el texto.

```python
TOTAL=0

lista_notas1=notas1.split(',')
lista_notas2=notas2.split(',')
lista_nom=prolijo(alumnos)
```

Separamos los archivos provistos por la cátedra con la función split, y invocamos la función "prolijo" que "limpia" el texto de nombres sin comas ni comillas.

```python
dicci={}

for indec in (range(len(lista_nom))):
    acumulado = int(lista_notas1[indec]) + int(lista_notas2[indec])
    dicci[lista_nom[indec]]=acumulado
    TOTAL+=acumulado
```

Creamos el diccionario donde guardaremos nuestra información. Recorremos las listas de notas sumando ambas notas para cada alumnos y agregando en cada caso un elemento al diccionario. En cada iteración vamos acumulando el total para el calculo del promedio según requiere el ejercicio.

```python
promedio=TOTAL/(len(lista_nom))
print("Los alumnos que obtuvieron puntaje por debajo del promedio son:")
for elem in dicci:
    if (dicci[elem]<promedio):
        print(elem)
```

Teniendo el acumulado de las notas en total, calculamos el promedio dividiendolo por la cantidad de alumnos, ya que el ejericicio pide el promedio de las notas totales, es decir de la suma de las dos notas de cada alumno. 

Por último se recorre el diccionario con un condicional que verifica que claves del diccionario, es decir alumnos, no alcanzan el promedio de las notas, imprimiendo sus nombres.

```python
print(f"El promedio de las notas es:{promedio}")
```
